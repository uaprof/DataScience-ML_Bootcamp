## Training & Test Dataset Split

## Multivariable Regression

```python
regr = LinearRegression()
regr.fit(X_train, y_train)

print('R^2 for train data is: ', regr.score(X_train, y_train))
print('R^2 for test data is: ', regr.score(X_test, y_test))
print('Intercept', regr.intercept_)

print(pd.DataFrame(data=regr.coef_, index=X_train.columns, columns=['coef'] ))
```

$$ VIF _{TAX} = \frac{1}{(1-R _{TAX} ^2)}$$

```python
variance_inflation_factor(exog=X_incl_const.values, exog_idx=1)
type(X_incl_const)
```

```python
vif = []
for i in range(X_incl_const.shape[1]):
    vif.append(variance_inflation_factor(exog=X_incl_const.values, exog_idx=i))
#print(vif)
```

## Model simplification & the BIC

```python
#Original model with log prices and alll fatures

X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

org_coef = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})

print('BIC is:', results.bic)
print('R squared is:', results.rsquared)
```

```python
# Reduced model #1 excluding INDUS

X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS'], axis=1)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

coef_minus_indus = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})

print('BIC is:', results.bic)
print('R squared is:', results.rsquared)
```

```python
# Reduced model #2 excluding AGE

X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS', 'AGE'], axis=1)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

coef_minus_age = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})

print('BIC is:', results.bic)
print('R squared is:', results.rsquared)
```

## Residuals & Residual Plot

```python
# Modified model: transformed (used log prices) & simplified (dropping two features)
prices = np.log(data['PRICE'])
features = data.drop(['PRICE','INDUS', 'AGE'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state = 10)

# Using Statsmodel
X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)
results = model.fit()

# Rsiduals
#residuals = y_train - results.fittedvalues
#results.resid

# Graph of Actual vs. Predicted Prices

corr = y_train.corr(results.fittedvalues)

plt.figure(figsize=(10,6))
plt.scatter(x=y_train, y=results.fittedvalues, c='navy', alpha=0.5)
plt.plot(y_train, y_train, color='cyan')

plt.xlabel('Actual log prices $y _i$', fontsize=14)
plt.ylabel('Predicted log prices $\hat y _i$', fontsize=14)
plt.title(f'Actual vs Predicted log prices: $y _i$ vs $\hat y _i$ (Corr {round(corr,2)})')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=np.e**y_train, y=np.e**results.fittedvalues, c='blue', alpha=0.5)
plt.plot(np.e**y_train, np.e**y_train, color='cyan')

plt.xlabel('Actual prices $y _i$', fontsize=14)
plt.ylabel('Predicted prices $\hat y _i$', fontsize=14)
plt.title(f'Actual vs Predicted prices: $y _i$ vs $\hat y _i$ (Corr {round(corr,2)})')
plt.show()

# Residuals vs Predicted values
plt.figure(figsize=(10,6))
plt.scatter(x=results.fittedvalues, y=np.e**results.resid, c='navy', alpha=0.5)


plt.xlabel('Predicted log prices $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted Values')
plt.show()

# Mean Squared Error
reduced_log_mse = round(results.mse_resid, 3)
reduced_log_rsquared = round(results.rsquared, 3)

```

```python
# Distribution of Residuals (log prices) - checking for normailty
resid_mean = round(results.resid.mean(), 3)
resid_skew = round(results.resid.skew(), 3)

plt.figure(figsize=(10,6))
sns.distplot(results.resid, color = 'navy')
plt.title(f'Log price model: residuals Skew ({resid_skew}), Mean ({resid_mean})')
plt.show()
```

```python
# Original model: norm prices all features
prices = data['PRICE']
features = data.drop(['PRICE'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state = 10)

# Using Statsmodel
X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)
results = model.fit()

# Graph of Actual vs. Predicted Prices

corr = y_train.corr(results.fittedvalues)

plt.figure(figsize=(10,6))
plt.scatter(x=y_train, y=results.fittedvalues, c='indigo', alpha=0.5)
plt.plot(y_train, y_train, color='cyan')

plt.xlabel('Actual prices $y _i$', fontsize=14)
plt.ylabel('Predicted prices $\hat y _i$', fontsize=14)
plt.title(f'Actual vs Predicted prices: $y _i$ vs $\hat y _i$ (Corr {round(corr,2)})')
plt.show()

#Graph of Predited values vs Residuals
plt.figure(figsize=(10,6))
plt.scatter(x=results.fittedvalues, y=results.resid, c='indigo', alpha=0.5)


plt.xlabel('Predicted prices $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted Values')
plt.show()

# Distribution of Residuals (Normal prices) - checking for normailty

plt.figure(figsize=(10,6))

resid_mean = round(results.resid.mean(), 3)
resid_skew = round(results.resid.skew(), 3)

plt.figure(figsize=(10,6))
sns.distplot(results.resid, color = 'indigo')
plt.title(f'Log price model: residuals Skew ({resid_skew}), Mean ({resid_mean})')
plt.show()

# Mean Squared Error
full_normal_mse = round(results.mse_resid, 3)
full_normal_rsquared = round(results.rsquared, 3)
```

```python
# Model Omitting Ket Features using log prices
prices = np.log(data['PRICE'])
features = data.drop(['PRICE', 'INDUS', 'AGE', 'LSTAT', 'RM', 'NOX', 'CRIM'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state = 10)

# Using Statsmodel
X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)
results = model.fit()

# Graph of Actual vs. Predicted Prices

corr = y_train.corr(results.fittedvalues)

plt.figure(figsize=(10,6))
plt.scatter(x=y_train, y=results.fittedvalues, c='#e74c3c', alpha=0.5)
plt.plot(y_train, y_train, color='cyan')

plt.xlabel('Actual log prices $y _i$', fontsize=14)
plt.ylabel('Predicted log prices $\hat y _i$', fontsize=14)
plt.title(f'Actual log vs Predicted prices with ommited falues: $y _i$ vs $\hat y _i$ (Corr {round(corr,2)})')
plt.show()

#Graph of Predited values vs Residuals
plt.figure(figsize=(10,6))
plt.scatter(x=results.fittedvalues, y=results.resid, c='#e74c3c', alpha=0.5)


plt.xlabel('Predicted prices $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted Values')
plt.show()

# Distribution of Residuals (Normal prices) - checking for normailty

plt.figure(figsize=(10,6))

resid_mean = round(results.resid.mean(), 3)
resid_skew = round(results.resid.skew(), 3)

plt.figure(figsize=(10,6))
sns.distplot(results.resid, color = '#e74c3c')
plt.title(f'Log price model: residuals Skew ({resid_skew}), Mean ({resid_mean})')
plt.show()

# Mean Squared Error
omitted_var_mse = round(results.mse_resid, 3)
omitted_var_rsquared = round(results.rsquared, 3)
```

```python
pd.DataFrame({'R-Squared': [reduced_log_rsquared, full_normal_rsquared, omitted_var_rsquared],
             'MSE':[reduced_log_mse, full_normal_mse, omitted_var_mse],
             'RMSE': np.sqrt([reduced_log_mse, full_normal_mse, omitted_var_mse])},
            index=['Reduced Log Model', 'Full Normal Price Model', 'Omitted Var Model'])
```
